# Soil Chemistry Workflow

This notebook scaffold supports the article **Soil Chemistry, Nutrient Cycles, and Land Systems**. It uses synthetic data for teaching soil organic carbon stock, base saturation, and nutrient-screening logic.

```python
import pandas as pd
from pathlib import Path

article_dir = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
soil = pd.read_csv(article_dir / 'data' / 'soil_monitoring_synthetic.csv')
soil['soc_stock_Mg_ha'] = soil['soil_organic_carbon_percent'] * soil['bulk_density_g_cm3'] * soil['depth_cm']
soil['base_saturation_percent'] = 100 * soil['base_cations_cmolc_kg'] / soil['cec_cmolc_kg']
soil[['site', 'land_use', 'pH', 'soil_organic_carbon_percent', 'soc_stock_Mg_ha', 'base_saturation_percent']].head()
```

```python
soil['phosphorus_flag'] = soil['phosphorus_mg_kg'].apply(lambda x: 'high_attention' if x > 60 else 'not_high')
soil['nitrate_flag'] = soil['nitrate_mg_kg'].apply(lambda x: 'high_attention' if x > 30 else 'not_high')
soil[['site', 'phosphorus_mg_kg', 'phosphorus_flag', 'nitrate_mg_kg', 'nitrate_flag']]
```
