# Astrochemistry Workflow

This notebook scaffold supports the article **Astrochemistry and the Molecular Universe**. It uses synthetic data for teaching molecular line matching, fractional abundance, Doppler velocity, thermal desorption, and photochemical screening.

```python
import math
import pandas as pd
from pathlib import Path

article_dir = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
data = pd.read_csv(article_dir / 'data' / 'astrochemical_survey_synthetic.csv')
c_km_s = 299792.458
data['radial_velocity_km_s'] = -c_km_s * (data['observed_frequency_GHz'] - data['rest_frequency_GHz']) / data['rest_frequency_GHz']
data['fractional_abundance'] = data['column_density_cm2'] / data['H2_column_density_cm2']
data[['source', 'environment', 'candidate_species', 'radial_velocity_km_s', 'fractional_abundance']].head()
```

```python
attempt_frequency = 1e12
data['desorption_rate_s1_simplified'] = attempt_frequency * [
    math.exp(-eb / temp) for eb, temp in zip(data['binding_energy_K'], data['dust_temperature_K'])
]
data[['source', 'candidate_species', 'dust_temperature_K', 'binding_energy_K', 'desorption_rate_s1_simplified']]
```
