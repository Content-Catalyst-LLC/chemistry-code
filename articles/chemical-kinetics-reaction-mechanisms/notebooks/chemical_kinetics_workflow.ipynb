# Chemical Kinetics and Reaction Mechanisms

Minimal notebook scaffold for integrated rate laws, Arrhenius analysis, reaction mechanism ODEs, and enzyme kinetics.

This notebook uses simplified educational data. It is not a validated kinetic model, safety method, industrial process tool, or regulatory workflow.

```python
import numpy as np
import pandas as pd

data = pd.read_csv('../data/first_order_data.csv')
subset = data[data['experiment'] == 'first_order_demo'].copy()
subset['ln_concentration'] = np.log(subset['concentration_mol_l'])
slope, intercept = np.polyfit(subset['time_min'], subset['ln_concentration'], deg=1)
k = -slope
half_life = np.log(2) / k
k, half_life
```
