# Ocean Carbonate Chemistry Workflow

This notebook scaffold supports the article **Ocean Chemistry and the Carbonate System**. It uses synthetic data for teaching carbonate speciation, aragonite saturation, and ocean carbon monitoring structure.

```python
import math
import pandas as pd
from pathlib import Path

article_dir = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
data = pd.read_csv(article_dir / 'data' / 'ocean_carbonate_monitoring_synthetic.csv')

K1 = 10 ** -6.0
K2 = 10 ** -9.1
KSP_ARAGONITE = 6.5e-7

def alpha2(pH):
    H = 10 ** (-pH)
    denominator = H**2 + K1 * H + K1 * K2
    return K1 * K2 / denominator

data['alpha_CO3'] = data['pH_total_scale'].apply(alpha2)
data['carbonate_umol_kg'] = data['alpha_CO3'] * data['DIC_umol_kg']
data['omega_aragonite_simplified'] = ((data['calcium_mmol_kg'] * 1e-3) * (data['carbonate_umol_kg'] * 1e-6)) / KSP_ARAGONITE
data[['station', 'water_type', 'pH_total_scale', 'carbonate_umol_kg', 'omega_aragonite_simplified']]
```
