```python
import numpy as np
import pandas as pd

params = pd.read_csv('../data/pair_potential_parameters.csv')
row = params.iloc[0]
r = np.linspace(row['sigma_angstrom'] * 0.85, row['sigma_angstrom'] * 3.0, 100)
u = 4 * row['epsilon_kj_mol'] * ((row['sigma_angstrom'] / r) ** 12 - (row['sigma_angstrom'] / r) ** 6)
pd.DataFrame({'r_angstrom': r, 'u_kj_mol': u}).head()
```
