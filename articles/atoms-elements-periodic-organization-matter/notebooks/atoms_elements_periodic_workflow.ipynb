This notebook uses simplified educational data. It is not a laboratory protocol, reference-data authority, safety procedure, regulatory method, or validated chemical workflow.

```python
from pathlib import Path
import pandas as pd

article_dir = Path('..')
isotopes = pd.read_csv(article_dir / 'data' / 'isotopes_sample.csv')
isotopes['neutron_number'] = isotopes['mass_number'] - isotopes['atomic_number']
isotopes['weighted_contribution_u'] = isotopes['isotopic_mass_u'] * isotopes['fractional_abundance']
isotopes
```

```python
weighted = isotopes.groupby('element_symbol', as_index=False).agg(
    weighted_atomic_mass_u=('weighted_contribution_u', 'sum'),
    abundance_sum=('fractional_abundance', 'sum'),
    isotope_count=('isotope', 'count')
)
weighted
```
