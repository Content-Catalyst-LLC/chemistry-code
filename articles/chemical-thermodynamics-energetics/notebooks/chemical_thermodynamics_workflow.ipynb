```python
import math
import pandas as pd

R = 8.314462618
gibbs = pd.read_csv('../data/gibbs_examples.csv')
gibbs['delta_g_standard_kj_mol'] = gibbs['delta_h_kj_mol'] - gibbs['temperature_k'] * gibbs['delta_s_j_mol_k'] / 1000
gibbs['K'] = gibbs.apply(lambda row: math.exp(-(row['delta_g_standard_kj_mol'] * 1000) / (R * row['temperature_k'])), axis=1)
gibbs
```
