# Water Chemistry Workflow

This notebook scaffold supports the article **Water Chemistry and Environmental Monitoring**. It uses synthetic data for teaching water-quality screening, nutrient-load estimation, and field-parameter interpretation.

```python
import pandas as pd
from pathlib import Path

article_dir = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
data = pd.read_csv(article_dir / 'data' / 'water_quality_monitoring_synthetic.csv')
data['ratio_to_benchmark'] = data['concentration'] / data['benchmark']
data[['site', 'medium', 'analyte', 'concentration', 'unit', 'benchmark', 'ratio_to_benchmark']].head()
```

```python
nutrients = data[data['analyte'].isin(['nitrate_as_N', 'phosphate_as_P', 'total_phosphorus', 'ammonia_as_N'])].copy()
nutrients = nutrients[nutrients['unit'] == 'mg/L']
nutrients['load_kg_day'] = nutrients['concentration'] * nutrients['flow_L_s'] * 0.0864
nutrients[['site', 'analyte', 'concentration', 'flow_L_s', 'load_kg_day']]
```
