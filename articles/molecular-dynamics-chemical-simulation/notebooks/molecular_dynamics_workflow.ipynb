```python
position = 0.0
velocity = 0.05
mass = 1.0
force = 0.10
dt = 0.5

acceleration = force / mass
new_position = position + velocity * dt + 0.5 * acceleration * dt**2
new_velocity = velocity + acceleration * dt

new_position, new_velocity
```
